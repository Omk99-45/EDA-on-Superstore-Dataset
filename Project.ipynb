### 2.BiVariate Analysis

```python
# 1.Display Region wise 1. Total sales  2.Total Profit

#Context :'Region','Sales'

#1.Total sales
df.groupby('Region')['Sales'].sum().round(2)
```

```python
# 2.Display % total sales

np.round(df.groupby('Region')['Sales'].sum().round(2) / df['Sales'].sum().round()*100,2)
```

```python
# 3.Total region wise profit
df.groupby('Region')['Profit'].sum().round(2)
```

```python
# 4 .Rrejion wise profit contribution.
np.round(df.groupby('Region')['Profit'].sum().round(2)/df['Profit'].sum().round()*100,2)
```

```python
# 5.Segment wise 
#  segment wise total sales

df.groupby('Segment')['Sales'].sum().round(2)
```

```python
# segment wise Total profit
df.groupby('Segment')['Profit'].sum().round(2)
```

```python
# 6.Display top 5 citys with max sales
(df.groupby('City')['Sales'].sum())
```

```python
# 7.Display region wise total sales of each segment

df.groupby(['Region','Segment'])['Sales'].sum().unstack()
```
